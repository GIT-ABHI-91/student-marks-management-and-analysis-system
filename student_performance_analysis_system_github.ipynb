```python
## setup for the database and the storage table
#creating and connecting with student_data database
# NOTE: Before running, replace 'host_info', 'user_info', 'user_password' with your MySQL credentials.
import mysql.connector as mc
from sqlalchemy import create_engine
import pandas as pd
class db_operation:
    def setup(self):
        self.temp_conn=mc.connect(      # making connection with mysql
            host='host_info',
            user='user_info',   # add your sql username here
            password='user_password'     #add your mysql password here
        )

        self.temp_curs=self.temp_conn.cursor()   #creating mysql pointer

        self.temp_curs.execute(" show databases like 'student_data' ")
        if self.temp_curs.fetchone() is None:   #checking if database exists or not
            self.temp_curs.execute(" create database student_data ")  #creating database in case of it's absence
            print("database created with name student_data")
        else:
            print("database found\ninitiating connection")
        self.temp_curs.close()
        self.temp_conn.close()

# creating and connecting with student table in student_data database
        self.conn=mc.connect(
            host='host_info',
            user='user_info',   # add your sql username here
            password='user_password',     #add your mysql password here
            database='student_data'
        )
        self.conn.autocommit = True

        self.curs=self.conn.cursor()
        self.curs.execute(" show tables like 'student' ")
        if self.curs.fetchone() is None:
            self.curs.execute(''' 
            create table student(
            id int auto_increment primary key,
            roll_no int unique,
            name varchar(100),
            dbms_marks int,
            dsa_marks int,
            oopj_marks int
            )''')
            print("storage table created and connected with")
        else:
            print("table storage found, can initiate program")

    def insert(self):   #method for data insertion
        self.name=input("ENTER STUDENT NAME- ")
        self.roll=int(input("ENTER ROLL NUMBER- "))
        self.dbms=int(input("ENTER DBMS MARKS- "))
        self.dsa=int(input("ENTER DSA MARKS- "))
        self.oopj=int(input("ENTER OOPJ MARKS- "))
        sql= " INSERT INTO student(roll_no,name,dbms_marks,dsa_marks,oopj_marks) VALUES (%s,%s,%s,%s,%s)  "
        values=(self.roll,self.name,self.dbms,self.dsa,self.oopj)
        self.curs.execute(sql,values)
        self.conn.commit()
        
        print("RECORD ENTERED")
        print("\n")

    def view(self):  # method for data to be viewed in form of dataframe
        # Replace the below string with actual connection URI
        engine = create_engine("mysql+pymysql://user:password@localhost/student_data")   #trying to connect mysql using sqlalchemy and encoding special characters in my password for better connectability
        self.df= pd.read_sql("select * from student",engine)
        print(self.df)
        print("\n")

    def update(self):   #method for updating student record
        roll=int(input("ENTER STUDENT ROLL NUMBER- "))
        sub=input("ENTER SUBJECT TO BE UPDATED")
        upd=int(input("ENTER UPDATED MARKS FOR THE SUBJECT- "))
        sql=f"update student set {sub} = %s where roll_no = %s"    # used string formatting here as %s placeholder is not supposed to hold dynamic values, it is only meant for storing up numeric values
        value=(upd,roll)
        self.curs.execute(sql,value)
        print("STUDENT DATA UPDATED")
        print("\n")

    def delete(self):   #method to delete a student record
        roll=int(input("ENTER STUDENT ROLL NUMBER- "))
        sql="delete from student where roll_no=%s"
        val=(roll,)
        self.curs.execute(sql,val)
        print("DATA OF STUDENT DELETED")
        print("\n")

    def view_spec(self):    #method to view the record of a particular student
        roll=int(input("ENTER STUDENT ROLL NUMBER- "))
        val=(roll,)
        sql="SELECT * from student where roll_no=%s"
        self.curs.execute(sql,val)
        result=self.curs.fetchall()
        print(f" ID:{result[0][0]} | ROLL NO.:{result[0][1]} | NAME:{result[0][2]} | DBMS MARKS:{result[0][3]} | DSA MARKS:{result[0][4]} | OOPJ MARKS:{result[0][5]}")
        print("\n")

    def close_conn(self):  # method for closing the database connections after leaving program
        self.curs.close()
        self.conn.close()


sq= db_operation()
sq.setup()   # to initiate database setup
print("welcome user")
print("1)INSERT STUDENT DATA\n2)VIEW STUDENT DATA\n3)UPDATE STUDENT DATA\n4)DELETE STUDENT DATA\n5)VIEW PARTICULAR RECORD\n")
ch=1
while(ch==1):   #loop based menu 
    op=int(input("ENTER OPERATION CODE- "))
    if op==1:
        sq.insert()
    elif op==2:
        sq.view()
    elif op==3:
        sq.update()
    elif op==4:
        sq.delete()
    elif op==5:
        sq.view_spec()
    else:
        print("INVALID OPERATION CODE")
    ch=int(input("WOUL YOU LIKE TO CONTINUE?(PRESS 1), OR EXIT(PRESS ANY OTHER KEY)- "))
print("EXITING THE PROGRGAM \nDISCONNECTING WITH DATABASE")
sq.close_conn()
```
